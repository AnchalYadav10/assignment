```python
def pass_test():
    p = input('Enter Password = ')
    count_u = []
    count_l = []
    count_s = []
    count_d = []
    if len(p)>=10:
        for i in p:
            if i.isupper():
                count_u.append(i)
            elif i.islower():
                count_l.append(i)       
            elif i.isdigit():
                count_d.append(i)
            else:
                count_s.append(i)
        if  len(count_u)>1 and len(count_l)>1 and len(count_s)>2 and len(count_d)>0:
            print('Valid Password')
        else:
            print('Invalid Password')
    else:
        print('Invalid Password')
pass_test()
```
