```python
#Q.1
# 'def' keyword is used to create a function in python.

def odd_num():
    l=[]
    for i in range(1,26):      
        if i%2!= 0:
            l.append(i)
    return l
print('Odd numbers in the range of 1 to 25 is {}'.format(odd_num()))
```

```python
#Q.2
# '*args' is used to pass n number of arguments on which operation of tuple can be performed
# '**args' is used to pass n number of argumrents on which operations of dictionary can be performed.


def greeting(arg,*argu):
    print('first word : ',arg)
    for args in argu:
        print('argument to be execute through *argu : ', args)
greeting('Have','a','nice','day')
```

```python
#Q.4
# A generator function allows you to create a custom iterator.A generator function returns a generator object that can be used to produce a sequence of values.
# A generator function is defined using the yield keyword. The yield keyword allows you to return a value from the function and then resume execution from that point the next time the function is called.
def test_fib1(n):
    a,b=0,1
    for i in range(n):
        yield a #yeild work to throw data and it become generator function
        a,b = b, a+b
fib=test_fib1(10)
n=10
for i in test_fib1(10):
    print(i)
```

```python
#Q.5
def prime():
    for n in range(2,1000):
        for i in range(2,n):
            if n%i==0: #since prime no is divisible by 1 and itself
                break
        else:
                yield n
a=iter(list(prime()))
i=1
while i<21:
    print(next(a),end=' ')
    i=i+1
```

```python
#Q.6 
n = int(input('enter your number = '))
def fibo():
    a,b=0,1
    while True:
        yield a  #yeild is used to throw the output which work similar to print
        a,b=b,a+b
fib=fibo()
for i in range(n):
    print(next(fib))
    
```
